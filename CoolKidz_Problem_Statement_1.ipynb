# Preprocessing

As there are too many video-text pairs and we have time constraints, we need to narrow them down before downloading the videos. Our method involves restricting the number of videos based on Frame Per Second. We can also filter out by removing clips with same captions, or remove clips with long duration. In our solution, we will only do filter for Frame per Second (FPS) and duplicated captions.

First, we remove videos with Frame Per Second lower than 25.

```python
import json

# Load the dataset from the JSON file
with open('mlda_data.json', 'r') as file:
    data = json.load(file)

# Function to filter out clips with fps less than 25
def filter_clips(data):
    filtered_data = {}
    for video_id, video_info in data.items():
        clips = video_info['clip']
        filtered_clips = {}
        for clip_id, clip_info in clips.items():
            if clip_info['fps'] >= 25:
                filtered_clips[clip_id] = clip_info
        if filtered_clips:  # Check if there are any clips left after filtering
            filtered_data[video_id] = video_info
            filtered_data[video_id]['clip'] = filtered_clips
    return filtered_data

# Filter the dataset
filtered_data = filter_clips(data)

# Save the filtered dataset to a new JSON file
with open('final.json', 'w') as file:
    json.dump(filtered_data, file, indent=4)

print("Filtered dataset saved to 'final.json'.")
```

We can remove them based on their captions. If there is more than one clip with the same caption, we will remove the rest, retaining only one unique clip. This can help in the diversification of the 10k dataset.

```python
with open('final.json', 'r') as file:
    data = json.load(file)

def remove_duplicate_clips(data):
    seen_captions = set()
    clips_to_retain = {}

    for video_id, video_content in data.items():
        for clip_file, clip_details in video_content['clip'].items():
            clip_has_unique_caption = True
            for scene in clip_details['scene_split']:
                caption = scene['caption']
                if caption in seen_captions:
                    clip_has_unique_caption = False
                    break  # Stop checking this clip if a duplicate caption is found
                seen_captions.add(caption)

            if clip_has_unique_caption:  # Only retain clips where all captions were unique at the time of checking
                if video_id not in clips_to_retain:
                    clips_to_retain[video_id] = {"url": video_content["url"], "clip": {}}
                clips_to_retain[video_id]['clip'][clip_file] = clip_details

    return clips_to_retain

# Remove duplicate clips
data_cleaned = remove_duplicate_clips(data)

# Save the cleaned data back to a file
with open('final.json', 'w') as file:
    json.dump(data_cleaned, file, indent=4)

print("Duplicates removed and cleaned file saved.")
```

#Extract images from videos without cutting the videos into clips

We found that there is a provided clip cutting python script. Before using the script, we need to download the videos first. After that, we proceed to extract the images from the clips. However, this method will take up a lot of memory usage. This was done in our first try but it is not efficient. (scripts in appendix)

Hence, we come out with a method by developing a script which can download the video as a temporary file, follow the time span of each clip and extract the frames in every second. Then delete the temporary video after it is done.

1. It iterates through each video clip in the JSON data.
2. For each clip, it extracts the URL, start time, end time, and scene splits.
3. It calculates the start and end times in HH:MM:SS format.
4. It computes the relative path based on the video ID and clip key.
5. It creates a new output folder based on the relative path within the new_root_dir.
6. It constructs the output filename for the extracted frame.
7. It calls the extract_frame_from_video function to extract the frame and save it to the specified output filename.

With this method, we do not need to use the cut_videos_mlda.py script, which is provided. This can also help to save our space as the video will be deleted after extraction completed, and no clips will be saved too.

```python
import yt_dlp as youtube_dl
import json
import subprocess
import os
from pathlib import Path

# Define the root directory for extracted frames
new_root_dir = './frames'

os.makedirs(new_root_dir, exist_ok=True)

def seconds_to_time(seconds):
    """Convert seconds to HH:MM:SS format."""
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return f"{int(hours):02d}:{int(minutes):02d}:{int(seconds):02d}"

def extract_frame_from_video(url, start_time, end_time, output_filename):
    """Download a segment of the video and extract a frame."""
    # Use yt_dlp to download the video segment
    ydl_opts = {
        'outtmpl': 'temp.mp4',
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/mp4',
    }
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

    # Convert start and end times to seconds
    start_seconds = sum(x * int(t) for x, t in zip([3600, 60, 1], start_time.split(":")))
    end_seconds = sum(x * int(t) for x, t in zip([3600, 60, 1], end_time.split(":")))

    # Extract the frame at the midpoint of the start and end times
    mid_time = (start_seconds + end_seconds) / 2
    mid_time_formatted = seconds_to_time(mid_time)

    # Use ffmpeg to extract the frame
    subprocess.run(['ffmpeg', '-i', 'temp.mp4', '-ss', mid_time_formatted, '-vframes', '1', output_filename])

    # Clean up the temporary video file using os.remove() for compatibility
    os.remove('temp.mp4')

# Load JSON data
with open('final.json') as f:
    data = json.load(f)

# Iterate through each video clip and its scene_splits
for video_id, video_info in data.items():
    url = video_info['url']
    for clip_key, clip_details in video_info['clip'].items():
        for scene in clip_details['scene_split']:
            start_time = scene['scene_cut'][0] / clip_details['fps']
            end_time = scene['scene_cut'][1] / clip_details['fps']
            start_time_formatted = seconds_to_time(start_time)
            end_time_formatted = seconds_to_time(end_time)

            # Compute the relative path from the original root directory
            relative_path = os.path.join(video_id, clip_key)
            # Compute the new output folder path
            new_output_folder = os.path.join(new_root_dir, relative_path)
            os.makedirs(new_output_folder, exist_ok=True)
            output_filename = os.path.join(new_output_folder, f"{scene['clip_id']}.jpg")

            extract_frame_from_video(url, start_time_formatted, end_time_formatted, output_filename)

print("Frame extraction to new root directory complete.")
```

#Frame-caption Pairing

After that, we want to match the images with their captions, according to the json file.

```python
import pandas as pd
import os
import json

# Function to read JSON data from a file
def read_json_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Path to the JSON file
json_file_path = './final.json'

# Read JSON data from the file
json_data = read_json_file(json_file_path)

base_path = './frames'

# Function to recursively list all image paths
def list_all_images(directory, image_paths):
    for entry in os.listdir(directory):
        full_path = os.path.join(directory, entry)
        if os.path.isdir(full_path):
            list_all_images(full_path, image_paths)
        elif full_path.endswith('.jpg'):
            image_paths.append(full_path)

# Initialize an empty list to hold all image paths
all_image_paths = []
list_all_images(base_path, all_image_paths)

# Function to process each image path and retrieve data
def process_image_paths(image_paths, json_data):
    data = []
    for path in image_paths:
        # Extracting video_id and clip_id from the path
        parts = path.split(os.sep)
        video_id = parts[-3]
        clip_id_with_suffix = parts[-2]  # includes '_frames' suffix
        clip_id = clip_id_with_suffix.replace('_frames', '')

        caption = "Caption not found"

        # Check if the video ID exists in the JSON data
        if video_id in json_data:
            for clip_key, clip_info in json_data[video_id]['clip'].items():
                for scene in clip_info['scene_split']:
                    if scene['clip_id'] == clip_id:
                        caption = scene['caption']
                        break  # Stop searching once the caption is found

        data.append({'Video ID': video_id, 'Clip ID': clip_id, 'Caption': caption, 'Frame Path': path})

    return pd.DataFrame(data)


# Process all image paths to generate DataFrame
df = process_image_paths(all_image_paths, json_data)

# Print out all image paths (Step 1)
for path in all_image_paths:
    print(path)

# Display the DataFrame (Steps 2 & 3)
print(df.head())

excel_file_path = './frame_data.xlsx'

# Save the DataFrame to an Excel file
df.to_excel(excel_file_path, index=False)

print(f"DataFrame saved to Excel file: {excel_file_path}")
```

# Rank Best Text-Video pairs

In order to rank the best Text_Video pairs, we adopted the CLIP model `ViT-B/32` which can encode Images and Text into vector representation. Once we obtain the vector representation, we could easily determine the quality of text-video pairs using cosine similarity.

```python
import torch
import clip
from PIL import Image
import pandas as pd

def get_clip_score(image_path, text):
    # Load the pre-trained CLIP model and the image
    model, preprocess = clip.load('ViT-B/32')
    image = Image.open(image_path)

    # Preprocess the image and tokenize the text
    image_input = preprocess(image).unsqueeze(0)
    text_input = clip.tokenize([text])

    # Move the inputs to GPU if available
    device = "cuda" if torch.cuda.is_available() else "cpu"
    image_input = image_input.to(device)
    text_input = text_input.to(device)
    model = model.to(device)

    # Generate embeddings for the image and text
    with torch.no_grad():
        image_features = model.encode_image(image_input)
        text_features = model.encode_text(text_input)

    # Normalize the features
    image_features = image_features / image_features.norm(dim=-1, keepdim=True)
    text_features = text_features / text_features.norm(dim=-1, keepdim=True)

    # Calculate the cosine similarity to get the CLIP score
    clip_score = torch.matmul(image_features, text_features.T).item()

    return clip_score
```
